```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

# Assignment 3
Part 1

```python
mpg_data = sns.load_dataset('mpg')

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
num_mpg = mpg_data.select_dtypes(include=numerics)
sns.heatmap(data=num_mpg.corr())
plt.show()
```

```python
sns.pairplot(data=num_mpg)
plt.show()
```

Part 2

```python
diamonds_data = sns.load_dataset('diamonds')
diamonds_data = diamonds_data.loc[~((diamonds_data['color']=='D')|(diamonds_data['color']=='E')|(diamonds_data['cut']=='Fair'))]
g = sns.FacetGrid(data=diamonds_data,col='cut',row='color')
g.map(sns.scatterplot,'price','carat')

for (i,j,k), data in g.facet_data():
    if data.empty:
        ax = g.facet_axis(i,j)
        ax.set_axis_off()
        
plt.show()
```

```python
crash_data = sns.load_dataset('car_crashes')
sns.lmplot(data=crash_data, x='speeding',y='total')
plt.show()
```

Part 4

```python
iris_data = sns.load_dataset('iris')
attrs = ['sepal_length','sepal_width','petal_length','petal_width']
fig, ax = plt.subplots(1,4,figsize=(20,8))

for i in range(0,4):
    sns.boxplot(x=iris_data['species'],y=iris_data[attrs[i]],ax=ax[i])
    
plt.show()
```
